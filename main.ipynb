# Random Walk

Suppose a drunk person is trying to walk in a "2d universe". Since he is drunk, he's unable to walk on a straight line. Hence, the probability of him taking a step further or a step back is the same.

Below is an extension of this situation, considering we have 10000 walkers, each taking 100 steps. When we see a map of this universe, showing where each walker stopped after this iterations, what will be the shape of the result data?


```python
import random
import matplotlib.pyplot as plt
```

```python
class Walker2D:
    """This walker wanders at random, back and forth. The decision is random,
    but luckly he has a GPS to see his position."""
    def __init__(self):
        self._pos = 0

    def _decide(self):
        return -1 if random.randint(0,1) == 0 else 1

    def take_step(self):
        self._pos += (self._decide())

    @property
    def gps(self):
        return self._pos

```

```python
no_of_walkers = 10000
no_of_steps = 100

gps_results = {}

for _ in range(no_of_walkers):
    walker = Walker2D()
    [walker.take_step() for _ in range(no_of_steps)]
    try:
        gps_results[walker.gps] += 1
    except KeyError:
        gps_results[walker.gps] = 1
```

```python
D = dict(sorted(gps_results.items()))
plt.bar(range(len(D)), list(D.values()), align='center')
plt.xticks(range(len(D)), list(D.keys()))
plt.show()
```
